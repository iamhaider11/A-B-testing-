```python
ad_group = df[df['test group'] == 'ad']
psa_group = df[df['test group'] == 'psa']
```

```python
conversion_rate_ad = ad_group['converted'].mean()
conversion_rate_psa = psa_group['converted'].mean()
```

```python
print(ad_group['converted'].isnull().sum())
print(psa_group['converted'].isnull().sum())
```

```python
from scipy.stats import ttest_ind

ad_group['converted'] = pd.to_numeric(ad_group['converted'], errors='coerce')
psa_group['converted'] = pd.to_numeric(psa_group['converted'], errors='coerce')


ad_group = ad_group.dropna(subset=['converted'])
psa_group = psa_group.dropna(subset=['converted'])


t_stat, p_value = ttest_ind(ad_group['converted'], psa_group['converted'], equal_var=False)

print(f'Conversion Rate for Group A: {conversion_rate_ad}')
print(f'Conversion Rate for Group B: {conversion_rate_psa}')
print(f'Test Statistic: {t_stat}')
print(f'P-value: {p_value}')

if p_value < 0.05:
    print("The difference in conversion rates is statistically significant.")
    if conversion_rate_ad > conversion_rate_psa:
        print("Group A has a higher conversion rate.")
    else:
        print("Group B has a higher conversion rate.")
else:
    print("There is no statistically significant difference in conversion rates.")
```
